**42_Object Detection using Contours by WebCam**

```python
import cv2
import numpy as np
```

```python
#Scaling
def nothing(x):
    pass

# Window
cv2.namedWindow("Window")

# Trackbars
##For threshold
cv2.createTrackbar("th","Window",0,255,nothing)

##For lower
cv2.createTrackbar("lb","Window",0,255,nothing)
cv2.createTrackbar("lg","Window",0,255,nothing)
cv2.createTrackbar("lr","Window",0,255,nothing)

##For upper
cv2.createTrackbar("ub","Window",255,255,nothing)
cv2.createTrackbar("ug","Window",255,255,nothing)
cv2.createTrackbar("ur","Window",255,255,nothing)

# On Video
cap = cv2.VideoCapture(0)

##for continous running:
while cap.isOpened():
    #for live video
    #boolean data, frame
    r, frame = cap.read()
    if r == True:

        #Detect Object
        ##get threshold value
        thr = cv2.getTrackbarPos("th","Window")
        
        LB = cv2.getTrackbarPos("lb","Window")
        LG = cv2.getTrackbarPos("lg","Window")
        LR = cv2.getTrackbarPos("lr","Window")
        
        UB = cv2.getTrackbarPos("ub","Window")
        UG = cv2.getTrackbarPos("ug","Window")
        UR = cv2.getTrackbarPos("ur","Window")

        lower = np.array([LB,LG,LR])
        upper = np.array([UB,UG,UR])

            
        
        frame = cv2.flip(frame, 1)   # flip for y axis
        frame = cv2.resize(frame,(400,400))

        #Object detection using of hsv color channel
        hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)


         #Mask
        m = cv2.inRange(hsv,lower,upper)


        #Filter
        res = cv2.bitwise_and(frame,frame,mask=m)

        #Removing noise
        fr = cv2.bitwise_not(res)

        #Contour
        ##Thresholding Image
        r,thi = cv2.threshold(m,thr,255,cv2.THRESH_BINARY)
        cnt,hr = cv2.findContours(thi,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

        cv2.drawContours(frame,cnt,-1,(255,0,0),2)
        
        
        cv2.imshow("Thresholding Image", thi)
        cv2.imshow("Resultant",res)
        cv2.imshow("Mask",m)
        cv2.imshow("Video",frame)
        if cv2.waitKey(25) & 0xff == ord("q"):
            break

    
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
